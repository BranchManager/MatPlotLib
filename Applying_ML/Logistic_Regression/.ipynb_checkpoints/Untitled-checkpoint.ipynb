# Logistic Regression

```python
from matplotlib import pyplot
import numpy as np
from scipy import optimize

#use numpys on load text to read in the data
data = np.loadtxt("ex2data1.txt", delimiter=",")
X = data[:,0:2]#grab Everything from the from the columns specified after the comma which are
#0:2 which are the 2 columns from the first row 0
y = data[:,2] #this does similarly form above

#print(data.size)
m = X.T[0].size #data.size
#I was stuck here for a minute
#I forgoet that data.size would give me the total number of indicies in the matrix which 
# was screwing up my calculations later on

#print(X[0].size)
pyplot.scatter(X.T[0],X.T[1],c=y)


Theta = np.zeros((3,1)) #Create Theta as a n+1 x1 matrix. where n is 2 for the two columns in X
X = np.concatenate([np.ones((m,1)),X], axis=1)
#the above ine will give me the extra column of ones that
#I forgot about originally

#print(X)
#print(Theta)
#print(X.T[0])
#print(y)
#print(data)
```

## Implementing Sigmoid

```python
#This Function just calculates the sigmoid function part of the assignment
def sigmoid1(hxT):
    minus_Z = -hxT
    #print(minus_Z)
    
    gz = 1/(1+np.exp(minus_Z))
    #print(gz)
    return gz
hxT = np.dot(Theta.T,X.T)
sigmoid = sigmoid1(hxT)
print(sigmoid)
```

### Cost Function

```python
def CalculateCost(Theta,X,y):
    hxT = np.dot(Theta.T,X.T)
    sigmoid = sigmoid1(hxT)
    ######################Cost######################
    fstprt = -y*np.log(sigmoid)
    #print(y)
    #print(np.log(sigmoid))
    sndprt = (1-y)*np.log(1-sigmoid)
    
    #print("hello")
    #print(fstprt)
    #print(sndprt)
    
    cost = (1/m)*np.sum(fstprt-sndprt)
    #print(cost)
    #####################Grad########################
    #print(sigmoid-y)
    #print(X)
    grad = (1/m)*np.dot((sigmoid-y),X)
    
    #We compute the cost in this funciton using the cost equestion for logistic regression
    return cost, grad
CalculateCost(Theta,X,y)
```

### Optimization

```python
options = {'maxiter': 400}
res = optimize.minimize(CalculateCost,
                       Theta,
                       (X,y),
                       jac=True,
                       method='TNC',
                       options=options)
cost = res.fun
theta=res.x

print(cost)
print(theta)
```

```python
hxT = np.dot(theta.T,X.T)
sig = sigmoid1(hxT)

#this code comes from diberge utils.py

#We only need 2 points to define a line so we choose two endpoints
print(np.min(X[:,1])-2)
plot_x=np.array([np.min(X[:,1])-2,np.max(X[:,1])+2])
#then calculate the decision boundary line
plot_y = (-1. / theta[2])*(theta[1]*plot_x+theta[0])

pyplot.scatter(X.T[1],X.T[2],c=y)
pyplot.plot(plot_x,plot_y)
```

## Predict

```python
def predict(X,theta):
    hxt = np.dot(X,theta)
    sig = sigmoid1(hxt)
    #print(sig)
    p = (sig>0.5)
    #print(p)
    return(p)
    #    return 1
    #else:
    #    return 0
p = predict(X,theta)
print(np.mean(p==y)*100)
#predict was pretty self explanatory. I just get an array of true or false for admitted 
#or not. We compare that against y's 1's and 0's to see how accurate ti was.
```
