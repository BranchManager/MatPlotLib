```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from IPython.display import display

df = pd.read_csv('Final_Mens_open21.1.csv')
Mens_data = pd.read_csv('men_main_21.1.csv')

pd.set_option('display.max_columns',None)
display(df.head(10))

```

```python
df['Body Height in inch']=df['Body Height in inch'].round(0)
df.iloc[57645:57655]

All_heights = df[df['Body Height in inch'].notna()]
All_heights.iloc[57645:57655]
display(df[df['Body Height in inch']==1.0])
#[display(Mens_data.loc[x]) for x in df[df['Body Height in inch']==6.0].index]
#display(df.groupby(['Body Height in inch']).mean().head(50))
#display(df['Body Height in inch'].value_counts().head(60))
df_heights = df[df['Body Height in inch'].between(60,83)]
average_by_height = df_heights.groupby(['Body Height in inch'])[['points']].mean()
#display()

#average_by_height.plot(x='Body Height in inch', y='points')
display(average_by_height.sort_values(by="points"))
```

```python
average_by_height = average_by_height.reset_index()
average_by_height.plot.scatter(x='Body Height in inch',y='points')
```
